```python
import pandas as pd
import re

# Sample data


# Function to clean and standardize values
def clean_other_apps(value):
    value_lower = str(value).lower()

    # Replace 'no', 'none', or similar with "None to say"
    if re.search(r'\bno\b|\bnone\b|\bnothing\b', value_lower):
        return 'None to say'

    # Standardize similar words (add more if needed)
    replacements = {'olx': 'OLX', 'fb': 'Facebook', '‘': '', '-': ''}
    for key, replacement in replacements.items():
        value_lower = value_lower.replace(key, replacement)

    return value_lower.strip()

# Apply the cleaning function to the 'Other apps' column
survey['Other Apps'] = survey['Other Apps'].apply(clean_other_apps)


```

```python
survey['Other Apps'].fillna('None to say',inplace=True)
```

```python
# Rename the list to avoid naming conflict
apps_to_remove = ['facebook', 'ebay']

def other_apps(value):
    value_lower = str(value).lower()

    for word in apps_to_remove:
        if word.lower() in value_lower:
            return ''
    
    return value  # Return the original value if no match is found

# Assuming survey is your DataFrame
survey['Other Apps'] = survey['Other Apps'].apply(other_apps)
```

```python
lisst_experience=[]
for i in survey['Whatsapp shopping experience']:
    lisst_experience.extend(i.split(','))
lisst_experience
```

```python
from collections import Counter



# Use Counter to count the frequency of each word
word_frequency = Counter(lisst_experience)

# Convert Counter to a dictionary
frequency_dict = dict(word_frequency)

merged_dict = {}

for key, value in frequency_dict.items():
    normalized_key = key.strip().lower()  # Remove trailing spaces and convert to lowercase
    merged_dict[normalized_key] = merged_dict.get(normalized_key, 0) + value

# Display the merged dictionary
print(merged_dict)

```

```python
import matplotlib.pyplot as plt

# Define categories based on keywords
furniture_keywords = ['Chair', 'sofa', 'table']
Travel=['Suitcases', 'travel backpack']
electronics_keywords = ['Watch', 'Tablet', 'Phone', 'Tv', ]
other_essentials_keywords = ['Ironbox', 'hairdryer', 'Water heater', 'Water Filter']
# Add more categories as needed

# Initialize counters for each category
furniture_count = 0
electronics_count = 0
other_essentials_count = 0
other_count = 0

# Categorize items
for item in lisst_experience:
    normalized_item = item.strip().lower()
    
    if any(keyword in normalized_item for keyword in furniture_keywords):
        furniture_count += 1
    elif any(keyword in normalized_item for keyword in electronics_keywords):
        electronics_count += 1
    elif any(keyword in normalized_item for keyword in Travel):
        electronics_count += 1
    elif any(keyword in normalized_item for keyword in other_essentials_keywords):
        other_essentials_count += 1
    else:
        other_count += 1

# Create a dictionary for the category counts
category_counts = {
    'Furniture': furniture_count,
    'Electronics': electronics_count,
    'Appliances': other_essentials_count,
    'Apparels': other_count
}

# Plotting the pie chart
plt.pie(category_counts.values(), labels=category_counts.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Exchange Goods')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
plt.show()
```
