```python

def remove_dol(price):
    if pd.notna(price):
        result = str(price).strip("$")
        return result
    else:
        return None  # or any other value you prefer for NaN cases

# Assuming 'Price' is a column in merged_df
reviews_df['loss'] = reviews_df['loss'].apply(remove_dol)
```

```python
reviews_df
```

```python
reviews_df.dropna(how='all', inplace=True)
```

```python
reviews_df.drop(columns='Category', inplace=True)
```

```python
reviews_df.columns = ['Review Title', 'Loss', 'App', 'Category']
```

```python
reviews_df.isna().sum()
```

```python
reviews_df.groupby('Category').count()
```

```python
main={'payment':'fake payment scam','switch':'bait and switch scam','refund':'no refund scam','vehicle':'fake vehicle scam','customer':'customer service','ship':'shipment damage/issue','meet':'meet-up scam','delivery':'non-delivery','phish':'phishing scam','money':'fake payment scam','merchandise':'payment issue'}
reviews_df['Category'] = reviews_df['Category'].astype(str)

def replace_words(category):
    for key, value in main.items():
        if key in category:
            return value
    return category

# Apply the function to the 'Category' column
reviews_df['Category'] = reviews_df['Category'].apply(replace_words)
```

```python
reviews_df.sort_values(by='Category')
```

```python
import pandas as pd

# Assuming 'None' values in the 'Loss' column should be treated as NaN
reviews_df['Loss'] = pd.to_numeric(reviews_df['Loss'], errors='coerce')

# Replace NaN values with a specific integer value (e.g., 0)
reviews_df['Loss'] = reviews_df['Loss'].fillna(0).astype(int)



```

```python
reviews_df.groupby('Category')['Loss'].mean()
```

```python
reviews_df['Loss'] = reviews_df.groupby('Category')['Loss'].transform(lambda x: x.replace(0, x.mean()))
        
```
